### Graphical Analysis - 4 Graphs

```python
g = sns.catplot(
    data=carData.sort_values("Make").tail(2000), kind="bar",
    x="Make", y="MaxMPG", alpha=.6, height=6).set(title="MPG by Make")
```

This graph looks at which cars are getting the best MPG in general. It is broken down by make of the car. Clearly, they're all pretty similar except for the EV, Tesla. 

```python
sns.histplot(data=carData, x="ReliabilityRating", bins=15,kde=True).set(title="Histogram - Reliability")
```

Wanted to just look to see what the cars in general were rated based on reliability. Thought it would be a little more evenly distributed, given how many stereotypes there are about unreliable cars. Instead, it seems like just about everything is above a 4/5 on the score -- so maybe reliability won't be too much of a factor here.

```python
sns.boxplot(x="Make", y="ConsumerReviews",
            hue="Used/New", palette=["m", "g"],
            data=carData.sort_values("Make").tail(2000)).set(title="Box Plot - Review Numbers")
```

The goal of this one was to look closely at how many consumer reviews were given per make here. It wasn't as helpful as I expected, because there were massive error bars and the information wasn't that great. However, I do think a car having way more consumer reviews than others is meaningful information.

```python
g = sns.lmplot(
    data=carData.sort_values("Make").tail(2000),
    x="ConsumerRating", y="ReliabilityRating", hue="Make",
    height=10).set(title="Reliability vs Consumer Rating")
```

This graph plotted Reliability Rating vs Consumer Rating -- seeing if there was a correlation between the consumer being happy and the car being reliable. Since most of the data points are in the top right corner, I would say that there is a correlation.

```python
#Splitting the data into training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#The target will be ConsumerRating since we are most interested in what the end user thinks of the car.
target = newDf['ConsumerRating']
```

```python
#Identifying the features as everything except for the target variable
features = newDf.loc[:, newDf.columns != 'ConsumerRating']
```

```python
#Actually splitting the data
x_train, x_test, y_train, y_test = train_test_split(features, target)
```

```python
#Checking sizes and everything looks to be as expected
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
#Initializing the linear regression model
model = LinearRegression()
```

```python
#Fitting the model to the training data
model.fit(x_train,y_train)
```

```python
#Getting the prediction
prediction = model.predict(x_test)
```

```python
#Turning those results into a dataframe for easier readability
predictDf = pd.DataFrame(prediction, columns = ['Predicted Value'])
```

```python
predictDf
```

```python
#Checking the accuracy of the model
model.score(x_test,y_test)
```

```python
#Finding the most important features based on feature selection. I chose the Mutual Info instead of Correlation
#Feature selection
f_selector = SelectKBest(score_func=mutual_info_regression, k='all')
#Learning from Training data
f_selector.fit(x_train, y_train)
#Transforming the training data
x_train_fs = f_selector.transform(x_train)
#Transforming the testing data
x_test_fs = f_selector.transform(x_test)
#Creating the plot
plt.bar([i for i in range(len(f_selector.scores_))], f_selector.scores_)
plt.xlabel("Column Index")
plt.ylabel("Value - Mutual Info")
plt.title("Columns and their Respective Mutual Info Values")
plt.show()
```
